```python
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib
rc('font',**{'family':'monospace','monospace':['Computer Modern Typewriter']})
rc('text', usetex=True)
matplotlib.rcParams["text.latex.preamble"].append(r'\usepackage{xfrac}')
%matplotlib inline
import matplotlib.mlab as mlab
from matplotlib.dates import MonthLocator, WeekdayLocator, DateFormatter
import numpy as np
import pandas
```

```python
months = MonthLocator(range(1, 13), bymonthday=1, interval=2)
monthsFmt = DateFormatter("%b '%y")


wikipediaEdits = np.array([180, 197, 103, 113, 169, 165, 470, 520, 536, 464, 341, 2901, 1203, 310, 218])
wikipediaViews = np.array([242952, 215387, 218565, 200902, 143188, 1899056, 217030, 165439, 196481, 206768, 423937, 241601, 156471, 224401, 222864])
wikidataEdits = np.array([11851922, 8186554, 6712724, 11788026, 10607889, 9013469, 9953821, 9356282, 8563604, 9840729, 13168713, 10187147, 8000526, 10454846, 10414970])
wikidataViews = np.array([11909308, 15633259, 13864167, 10274106, 18073528, 10832818, 13091566, 11463824, 13443838, 12203616, 15859392, 23138442, 7588180, 16530174, 17247483])
dates = pandas.date_range('2015-07-01', '2016-09-30', freq='M')

assert(len(dates) == len(wikidataEdits))
```

```python
fig, ax = plt.subplots(1, 1)
ax.plot_date(dates, np.log(np.true_divide(wikidataEdits, wikidataViews)), 'r-', linewidth=2.0, label=r'$\mathtt{log(\sfrac{edits}{views})}$ on Wikidata')
ax.plot_date(dates, np.log(np.true_divide(wikipediaEdits, wikipediaViews)), 'g--', linewidth=2.0, label=r'$\mathtt{log(\sfrac{edits}{views})}$ on Wikipedia')
ax.xaxis.set_major_locator(months)
ax.xaxis.set_major_formatter(monthsFmt)
ax.autoscale_view()
ax.grid(True)


plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,
           ncol=1, mode="expand", borderaxespad=0.)
fig.autofmt_xdate()
plt.savefig('EditsViewsGraph.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
